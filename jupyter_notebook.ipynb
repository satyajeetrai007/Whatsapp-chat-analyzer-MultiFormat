```python
import re 
import pandas as pd
import datetime
```

```python
f = open('WhatsApp Chat with SAC NIT Meghalaya.txt','r',encoding ='utf-8')
```

```python
data = f.read()
```

```python
print(len(data.split('\n')))
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s\w{2}\s-\s'
```

```python
message = re.split(pattern,data)[1:]
```

```python
message
```

```python
len(message)
```

```python
dates = re.findall(pattern,data)
```

```python
dates
```

```python
df = pd.DataFrame({'messages':message,
                  'dates':dates
                  })
```

```python
df['dates']
```

```python

date_str_cleaned = df['dates'].replace('\u202f', ' ')

# Define the correct format string
date_format = "%d/%m/%y, %I:%M %p - "

# Convert to datetime
date_time_obj = pd.to_datetime(date_str_cleaned, format=date_format)
```

```python
df['dates'] = date_time_obj
```

```python
df.rename(columns={'dates':'date'},inplace = True)
```

```python
df
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['messages']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['users'] = users
df['message'] = messages
df.drop(columns=['messages'],inplace = True)
df
```

```python
df['users'].value_counts().head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['min'] = df['date'].dt.minute
```
